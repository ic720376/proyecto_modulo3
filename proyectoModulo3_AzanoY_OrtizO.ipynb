***PROYECTO***

```python
# Importar librerías
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

# Importamos función para integración numérica
from scipy.integrate import odeint
```

```python
def predator_prey(x,t):
    x1 = x[0]
    x2 = x[1]
    dx1 = 1.2*x1 - 0.6*x1*x2
    dx2 = -0.8*x2 + 0.3*x1*x2
    return np.array([dx1, dx2])
```

```python
#Condicion Inicial
x0 = np.array([2,1])
# Vector de tiempo
t = np.linspace(0, 30, 100)
```

```python
# Resolver numéricamente
xx = odeint(predator_prey, x0, t)
# Separamos soluciones
x1 = xx[:, 0]
x2= xx[:, 1]
```

```python
# Gráficas de x1 vs. t y x2 vs. t
plt.figure(figsize=(6,6))
plt.plot(t, x1, label="Prey")
plt.plot(t, x2, label= "Predator")
plt.xlabel("Tiempo $t$")
plt.axhline(y=0, color="k")
plt.axvline(x=0, color="k")
plt.legend(loc="best")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(x1, x2, label= "interaccion entre depredador y presa")
plt.plot(x1[0], x2[0], "*", ms=10, label = "Inicio de depredadores/presa")
plt.plot(x1[-1], x2[-1], "*", ms=10, label = "Final de depredadores/presa")
plt.xlabel("presa en el tiempo $x_1(t)$")
plt.ylabel("depredador en el tiempo $x_2(t)$")
plt.axhline(y=0, color="k")
plt.axvline(x=0, color="k")
plt.legend(loc = "best")
plt.grid()
plt.show()
```

```python
def clima(x,t):
    x1 = x[0]
    x2 = x[1]
    x3 = x[2]
    dx1 = -10*x1 + 10*x2
    dx2 = 28*x1 -x2 -x1*x3
    dx3 = -2.66667*x3 +(x1*x2)
    return np.array([dx1, dx2,dx3])
```

```python
x0 = np.array([5,5,5])
t = np.linspace(0, 20,100)
```

```python
xx = odeint(clima, x0, t)
x1 = xx[:, 0]
x2 = xx[:, 1]
x3 = xx[:,2]
```

```python
#Graficar las soluciones en el tiempo, es decir, theta vs. t y omega vs. t
plt.figure(figsize=(8,6))
plt.plot(t, x1,)
plt.plot(t, x2,)
plt.xlabel("Tiempo $t$")
plt.ylabel("$x$")
plt.axhline(y=0, color="k")
plt.axvline(x=0, color="k")
plt.legend(loc="best")
plt.grid()
plt.show()
```
